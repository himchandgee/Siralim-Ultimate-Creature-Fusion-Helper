```python
import csv
import os
from nltk.tokenize import word_tokenize

# Define the path to your CSV file
csv_file_path = 'Siralim Ultimate Creature and Trait Sheet - Creatures.csv'

# Define the directory where you want to save the text files
output_directory = 'text_files'
os.makedirs(output_directory, exist_ok=True)

# Open the CSV file for reading
with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # Iterate over each row in the CSV file
    for row in reader:
        # Use the 'name' column to generate a filename (make sure it's a valid filename!)
        # You might need to replace invalid characters in the name
        filename = f"{row['Name'].replace('/', '_')}.txt"
        file_path = os.path.join(output_directory, filename)
        
        # Tokenize the text content you want to write to the file using NLTK
        # Assuming the content is in a column named 'content'
        tokens = word_tokenize(row['Trait Description'])
        text_content = ' '.join(tokens)
        
        # Open the text file and write the content
        with open(file_path, 'w', encoding='utf-8') as text_file:
            text_file.write(text_content)
        
        print(f"File '{filename}' written.")

# Note: This script does simple tokenization for demonstration. 
# You might need to adjust the text processing according to your needs.
```

```python
directory = 'text_files'

# Load English stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stop_words(text):
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if not word.lower() in stop_words]
    return ' '.join(filtered_text)

# Process each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".txt"):  # Ensure processing only text files
        filepath = os.path.join(directory, filename)
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read()
            
            # Splitting content into groups based on comma
            if ',' in content:
                group_trigger, group_effects = content.split(',', 1)
                
                # Remove stop words
                cleaned_trigger = remove_stop_words(group_trigger)
                cleaned_effects = remove_stop_words(group_effects)
                
                # Optional: Write back to file or print
                print(f"File: {filename}")
                print("Cleaned Trigger:", cleaned_trigger)
                print("Cleaned Effects:", cleaned_effects)
                print()
```
